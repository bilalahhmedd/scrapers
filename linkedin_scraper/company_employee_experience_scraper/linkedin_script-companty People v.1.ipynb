```python
dfcomp = pd.read_excel(open('Companylist.xlsx','rb'))
dfcomp.columns = ['A', 'companies']
dfcomp=dfcomp.drop(columns=['A'])
dfcomp=dfcomp.drop(0)
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.linkedin.com/login")
sleep(2)
username = driver.find_element_by_xpath('//input[@name="session_key"]')
# input your username here
username.send_keys(user)
password = driver.find_element_by_xpath('//input[@name="session_password"]')
# input ur password here
password.send_keys(passw)
password.send_keys(Keys.RETURN)
sleep(2)
```

```python
def scroll():
    scrolling = True
    last_position = driver.execute_script("return window.pageYOffset;")

    while scrolling:

        scroll_attempt = 0
        while True:
            # check scroll position
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
            sleep(2)
            curr_position = driver.execute_script("return window.pageYOffset;")
            if last_position == curr_position:
                scroll_attempt += 1

                # end of scroll region
                if scroll_attempt >= 2:
                    scrolling = False
                    break
                else:
                    sleep(2) # attempt another scroll
            else:
                last_position = curr_position
                break
```

```python
for ind ,row in dfcomp.iterrows():
    comptemp= "".join(row['companies'].split())
    driver.get("https://www.linkedin.com/company/"+comptemp+"/people/")
    sleep(2)
    dfcomp.loc[ind,"searched url"]=comptemp

    try:
        c = driver.find_elements_by_class_name("artdeco-empty-state__message")
        if(len(c)>0):
            dfcomp.loc[ind,"status"]='false'
        else:
            dfcomp.loc[ind,"status"]='true'
            scroll()            
            AllEmp = driver.find_elements_by_class_name("org-people-profiles-module__profile-item")
            dfcomp.loc[ind,"count"]=len(AllEmp)
            if(len(AllEmp)>0):
                listuser=[]
                listdetails=[]
                listlink=[]
                for emp in AllEmp:
                    try:
                        user= emp.find_element_by_xpath('.//div[1]/section[1]/div[1]/div[1]/div[2]/div[1]').text
                    except:
                        user=""
                    try:
                        details= emp.find_element_by_xpath('.//div[1]/section[1]/div[1]/div[1]/div[2]/div[2]/div[1]').text
                    except:
                        details=''
                    try:
                        linkk= emp.find_element_by_xpath('.//div[1]/section[1]/div[1]/div[1]/div[1]/a')
                        link = linkk.get_attribute('href')
                    
                    except:
                        link=""
                    listuser.append(user)
                    listdetails.append(details)
                    listlink.append(link)
                
                Pt = [listuser,listdetails,listlink]
                df = pd.DataFrame (Pt).transpose()
                df.columns = ['User','Details',"Link"]
                df.to_csv("./linkedin-company-People/"+row['companies']+".csv")
    except:
        print("invalid")
                
```
