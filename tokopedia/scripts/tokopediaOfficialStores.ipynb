```python
searchlist=['electronics', 'fashion', 'baby', 'tour', 'travel', 'office','stationary', 'handphone','tablet']
```

All of Scraped Shop Urls will be stored In Temp Folder 

```python
Endcounter=1000 #number of stores
PreviousStores=[]
try:
    storedf=pd.read_csv("temp/MainStoreLink.csv")
    
    PreviousStores= list(storedf['StoreLink'])
except:
    print("Main list Not Found in Temp Folder")
```

```python
driver = webdriver.Chrome(options=opt)
```

```python
print("Chrome Driver Initiated")
print("Getting the list of products for the keyword")
```

```python
storelist=[]
breakk=False
```

```python
for search in searchlist:
    list2=[]
    driver.get("https://www.tokopedia.com/search?official=true&st=shop&q="+search)
    while True:
        sleep(3)
        while True:
            if(len(driver.find_elements_by_xpath(".//section[@class='css-1frsvni-unf-card eyk31ek0']"))>0):
                driver.refresh()
                sleep(3)



            last_height = driver.execute_script("return document.body.scrollHeight")
            while True:
                # Scroll down to bottom
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                # Wait to load page
                sleep(2)
                # Calculate new scroll height and compare with last scroll height
                new_height = driver.execute_script("return document.body.scrollHeight")
                if new_height == last_height:
                    break
                last_height = new_height
            sleep(5)
            if(len(driver.find_elements_by_xpath(".//div[@class='css-8f8se6-unf-dialog ek4e0y60']"))>0):
                driver.refresh()
                sleep(3)

            else:
                break





        Allshops=driver.find_elements_by_xpath(".//div[@data-testid='shop-card']")
        for shop in Allshops:
            shoplink=shop.find_element_by_xpath(".//a[@data-testid='shop-card-header']")
            shoplink=shoplink.get_attribute("href")

            if shoplink not in storelist:
                if shoplink not in PreviousStores:

                    storelist.append(shoplink)
                    PreviousStores.append(shoplink)
                    list2.append(shoplink)
                    print("store number"+str(len(storelist))) 
                    print(shoplink)
                    df=pd.DataFrame({'StoreLink':storelist})
                    df.to_csv("storelinks.csv")
                    df3=pd.DataFrame({'StoreLink':list2})
                    df3.to_csv("Finaloutput/"+search+"storelinks.csv")

                    df2=pd.DataFrame({'StoreLink':PreviousStores})
                    df2.to_csv("temp/MainStoreLink.csv")

                    if len(storelist)== Endcounter:
                        breakk=True
                        break



                else:
                    print("Store Allready Exists in Previously Scraped Stores")
            else:
                print("store already Scraped")

        if (breakk):
            break
        if(len(driver.find_elements_by_xpath(".//button[@class='css-1m1b4qs-unf-pagination-item e19tp72t3']"))>0):
            nextt=driver.find_element_by_xpath(".//button[@class='css-1m1b4qs-unf-pagination-item e19tp72t3']")
            desired_y = (nextt.size['height'] / 2) + nextt.location['y']
            window_h = driver.execute_script('return window.innerHeight')
            window_y = driver.execute_script('return window.pageYOffset')
            current_y = (window_h / 2) + window_y
            scroll_y_by = desired_y - current_y

            driver.execute_script("window.scrollBy(0, arguments[0]);", scroll_y_by)
            nextt.click()
        else:
            break

                
    
```

```python
print("getting All the product's prices of each shop")
# Productlist=[]
for index, row in df.iterrows():
    soldVolume=0
    
    
    if index>=startindex:
#     if(pd.isnull(row["SoldNumber"])):
        try:
            Productlist=[]
            pricelist=[]
            soldNumberList=[]

            print("Store  Number "+str(index))
            storelink=row['StoreLink']+"/product"
            pagenumber=1

            driver.get(storelink)

            while True:
                print("page#" + str(pagenumber))

                sleep(5)
                if pagenumber==1:
                    try:
                        try:
                            name=driver.find_element_by_xpath('.//h1[@class="css-1cwp34r"]').text
                            print(name)
    #                         df.at[index,"ShopRating"]=shopQuality
                            df.at[index,"ShopName"]=name



                            shopQuality=driver.find_element_by_xpath('.//h2[@class="css-3dui13-unf-heading-unf-heading e1qvo2ff2"]').text
                            print(shopQuality)
                            df.at[index,"ShopRating"]=shopQuality

                        except:
                            pass



                        try:
                            driver.find_element_by_xpath('.//button[@class="css-jbzeaf-unf-btn e1ggruw00"]')
                        except:
                            driver.refresh()

                        sleep(5)
                        driver.find_element_by_xpath('.//button[@class="css-jbzeaf-unf-btn e1ggruw00"]').click()
                        sleep(5)
                        shopinfo=driver.find_element_by_xpath('.//div[@class="css-v72a1a-unf-modal e1s17eky0"]')
                        description=shopinfo.find_elements_by_xpath('.//p[@class="css-1fe13z7-unf-heading-unf-heading e1qvo2ff8"]')
                        location=description[1].text
                        df.at[index,'Location']=location
                        description=description[0].text
                        df.at[index,'Description']=description
                        dateopen=shopinfo.find_element_by_xpath('.//p[@class="css-2in2o4-unf-heading-unf-heading e1qvo2ff8"]').text
                        df.at[index,'Join date']=dateopen

                        shopinfo.find_element_by_xpath(".//button").click()
                    except:
                        driver.refresh()

                        sleep(5)
                for x in range(0,2):
                    total_height = int(driver.execute_script("return document.body.scrollHeight"))

                    for i in range(1, total_height, 10):
                        driver.execute_script("window.scrollTo(0, {});".format(i))

                sleep(5)
                allproducts=driver.find_elements_by_xpath(".//div[@class='css-1sn1xa2']")
                print(len(allproducts))
                for product in allproducts:
                    try:
                        price=product.find_element_by_xpath(".//div[@class='css-rhd610']").text
#                         if "." not in price:
#                             price=price+".000"

        #                 price
        #                 tt=price
                        t=price
                        dec=float(t[-4:])
                        before=t[:-4]

                        floatt=float(''.join(x for x in before if x.isdigit()))
                        price=floatt+dec
                        pricelist.append(price)
                        soldnumber=0
                        if len(product.find_elements_by_xpath(".//span[@class='css-1kgbcz0']"))>0:

                            sold=product.find_element_by_xpath(".//span[@class='css-1kgbcz0']").text
                            sold
                            soldnumber=float(''.join(x for x in sold if x.isdigit()))
                        soldNumberList.append(soldnumber)
                        soldVolume=soldVolume+(price*soldnumber)
                    except:
                        pass
                if(len(driver.find_elements_by_xpath(".//a[@data-testid='btnShopProductPageNext']"))>0):
                    nextt=driver.find_element_by_xpath(".//a[@data-testid='btnShopProductPageNext']")
                    desired_y = (nextt.size['height'] / 2) + nextt.location['y']
                    window_h = driver.execute_script('return window.innerHeight')
                    window_y = driver.execute_script('return window.pageYOffset')
                    current_y = (window_h / 2) + window_y
                    scroll_y_by = desired_y - current_y

                    driver.execute_script("window.scrollBy(0, arguments[0]);", scroll_y_by)
                    nextt.click()
                    pagenumber+=1
                else:
                    break
            df.at[index,'SoldNumber']=",".join(map(str,soldNumberList))  

            df.at[index,'ProducPrices']=",".join(map(str, pricelist))

    #         df.at[index,'ProducNames']=",".join(namelist)
            df.at[index,'NumberOfProducts']=len(pricelist)
            df.at[index,'SoldVolume']=soldVolume


            pri=pricelist


            mean = sum(pri) / len(pri)
    #             print("54")
            df.at[index,'meanPrice']=mean
            variance = sum([((x - mean) ** 2) for x in pri]) / len(pri)
            sd = variance ** 0.5
            df.at[index,'StandardDeviation']=sd
    #             print("555")
            Sum = sum(pri)
            df.at[index,'PriceVolume']=Sum
    #             print("15545")
            minn=min(pri)
            df.at[index,'MinimumPrice']=minn
    #             print("5847877")
            df.at[index,'MaximumPrice']=max(pri)
    #             print("1dfsdfs")
    #             df.at[index,"ShopRating"]=shopQuality
    #             df.at[index,"ShopName"]=name

    #             df.at[index,'Location']=location

    #             df.at[index,'Description']=description
    #             df.at[index,'Join date']=dateopen



            df.to_csv("Finaloutput/AllshopMetadataCalculation.csv")

            
        except Exception as e:
            print(e)
            driver.close()
            sleep(2)
            opt = webdriver.ChromeOptions()
            opt.add_argument("--start-maximized")
            opt.add_experimental_option("excludeSwitches", ["disable-popup-blocking"])

            # opt.add_argument("--lang=en")

            chromedriver_autoinstaller.install()
            driver = webdriver.Chrome(options=opt)

            
```
